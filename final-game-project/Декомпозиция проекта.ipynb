## Декомпозиция проекта

## 1. Вводная часть

В этой части описаны:
- для кого это исследование
- зачем оно понадобилась
- какая основная цель исследования
- описание предосавленных данных
- план исследования

## 2. Обзор данных

- Импорт библиотек. 
- Чтение файлов с обработкой столбцов с датами, вывод первых 10 строк, проверка типов данных и пропусков в столбцах -  все одна функция.
- Написание общего вывода нужна ли преодбработка и если да, то почему.

## 3. Предобработка данных

### 3.1 Обработка пропусков

Исследование пропусков в столбцах `building_type` и `project_type` и их предообработка, если они не объясняются спецификой сбора данных

### 3.2 Обработка дубликатов

- Удаление явных дубликатов, если они есть.
- Поиск неявных дубликатов с подсчетом значений по столбцам `project_type`, `building_type` и `event` в датасете `game_actions`.
- Сравнение значени по столбцу `source` в датасете `user_source` и `ad_costs` для того, чтобы позже объединенить информацию без потерь

### 3.3 Исследование за какой период есть данные
- Линейный график о распределении времени `event_datetime` в датасете `game_actions`
- Минимальная и максимальная дата в столбце `day` в датасете `ad_costs`

### 3.4 Вывод

О предобработке и полученных результатах.

## 4. Создание функции для исследования

### 4.1 Функция для создания профилей игроков
Датафрейм `profiles`:
- `first_ts` - дата первого посещения
- `user_id` - id пользователя
- `source` - рекламны канал, с которого пришел пользователь
- `success` - успешностью прохождение 1-го уровня, булевое значение (True/False)
- `acquisition_cost` - расходы на привлечение 1-го пользователя, расчитанные как затраты на канал в день привлечение на число привлеченных пользователей

### 4.2 Функция для рассчета конверсии

С возможностью далее использовать ее не только для всех пользователей, но и с разбивкой по рекламным каналам.

### 4.3 Функция для визуализации конверсии
Линейый график для общей конверсии, а также с разбивкой по рекламным каналам.

## 5. Исследование рекламных каналов и их влияния на поведение пользователей

### 5.1 Динамика изменения расходов
Линейный график в общем и по рекламным каналам.

### 5.2 Распределение долей игроков по рекламным каналам
Интерактивная столбчатая диаграмма с долями завершивших игроков 1-ый уровень и тех, кто не прошел 1-ый уровень, по рекламным каналам

### 5.3 Cредняя стоимость привлечения 1-го пользователя по рекламным каналам

### 5.4 Визуализация конверсии
Вызов функции для отрисовки общего графика конверсии и по рекламным каналам во времени

### 5.5 Доли игроков, завершивших 1-ый уровень, по рекламным каналам
Интерактивная круговая диаграмма

### 5.6 Окупаемость по рекламным каналам
Визуализациия окупаемости по рекламным каналам. Рассчитывается как отношение количества построек игроков к стоимости их привлеечения во времени.

### 5.7 Среднее количество построек на 1-го пользователя по рекламным каналам
- Среднее количество строений у игроков, кто не прошел 1-ый уровень
- Среднее количество построек на 1-го пользователя по рекламным каналам
- Корреляция между количеством строений и прохождением

### 5.8 Типы строении по рекламным каналам
- На каком типе строении чаще всего пользователи останавливаются и не проходят 1-ый уровень до конца
- Столбчатая диаграма с распределением типов строенний по рекламным каналам

### 5.9 Корреляция между временем прохождения и успешным завершением 1-го уровня
Если есть положительная корреляция, то можно проверить среднее время прохождения первого уровня по рекламным каналам.

### 5.10 Топ-5 успешных стратегии прохождение 1-го уровня по рекламным каналам
Например, 3 `assembly_shop` + 5 `spaceport` или 2 `assembly_shop` + 3 `research_center`. Это поможет выявить откуда приходят опытные игроки.

### 6. Проверка статистических гипотез

### 6.1 Время завершения 1-го уровня различается в зависимости от способа прохождения
Для каждого игрока будет рассчитана разница между временем начала прохождения первого уровня и временем его окончания. После этого пользователи будут поделены на группы по способу прохождения и их значения будут сравниваться при помощи t-теста.

### 6.2 Поведение пользователей различается в зависимости от рекламного канала
Проверка гипотезы о разности долей событий в зависимости от рекламного канала с помощью z-теста с применение поправки на множественное сравнение. Это сможет показать то, что поведение пользователей различается в зависимости от рекламного канала или наоборот, что рекламные каналы не влияют на поведение пользователей.

## 7. Общий вывод
Написание общего вывода, включающий все предыдущие результаты, особенно если были найдены интересные закономерности и паттерны поведения пользователей. В выводе обязательно должен быть ответ на вопрос какой рекламный канал лучше всего для привлечения игроков и почему + описание оставшихся 3-х каналов с причиной почему они хуже.
